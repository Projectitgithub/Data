# Analysis and Visualization.

```python
# between current tabacoo user and whic area in present?
d['Area'].value_counts()
```

```python
plt.figure(figsize=(10,5))
d['Area'].value_counts().plot(kind="bar")
plt.show()
```

```python
d.groupby('Area')['Current tobacco users (%)'].sum()
```

```python
# Analysis for State/UT AND Current bidid users(%)?
d.groupby('State/UT')['Current bidi users (%)'].value_counts()[0:51]
```

```python
plt.figure(figsize=(10,5))
d['Current bidi users (%)'].value_counts().plot(kind="bar")
plt.show()
```

```python
# Using barplot for State/UT And'Current bidi users (%)'?
plt.bar(x='State/UT',height='Current bidi users (%)',data=d)
plt.xticks(rotation=90)
plt.show()
```

```python
#Analysis for State/UT AND Current smokeless tobacco users (%)?
d.groupby('State/UT')['Current smokeless tobacco users (%)'].value_counts()[0:51]
```

```python
# between the State/UT AND CURRENT SMOKELESS TOBACOO USERS? USING A BAR PLOT.
plt.bar(x='State/UT',height='Current smokeless tobacco users (%)',data=d)
plt.xticks(rotation=90)
plt.show()
```
